# 課題7.2

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 1.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()

fig.savefig("test.png")
plt.show()
```

# 課題 7.5

```python
# 課題7.4を関数化
def data_generation(num):
  rng = np.random.default_rng()
  input_data = rng.random(num)
  y = np.sin(2 * np.pi * input_data)
  noise = np.random.normal(loc=0, scale=0.1, size=num)
  new_y = y + noise
  return input_data, new_y

for num in [10,20,50,100]:
  input_data, new_y = data_generation(num)
  np.savetxt(f"{num}.data", np.column_stack((input_data, new_y)), fmt='%f %f')
```
